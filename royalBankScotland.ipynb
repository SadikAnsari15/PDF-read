```python
import camelot
import pandas as pd

# Extract tables from all pages using the 'stream' flavor
tables = camelot.read_pdf("RBS-statement-1.pdf", pages="all", flavor="stream")

# Combine all tables into a single DataFrame
dfs = [table.df for table in tables]
df = pd.concat(dfs, ignore_index=True)

# Drop any empty rows or columns if needed
df_cleaned = df.dropna(how='all', axis=0)
df_cleaned = df_cleaned.dropna(how='all', axis=1)  # Optionally drop empty columns

# Reset index
df_cleaned.reset_index(drop=True, inplace=True)

# Create separate columns for Date and Type
# We need to extract 'Date' from the first column when it starts with a date format
date_rows = df_cleaned[0].str.contains(r'\d{1,2} \w{3} \d{4}', na=False)

# Create a new DataFrame to store the structured data
structured_data = []

for index, row in df_cleaned.iterrows():
    if date_rows[index]:  # If the row contains a date
        date_type = row[0].split('\n')  # Split by newline to separate Date and Type
        date = date_type[1].strip() if len(date_type) > 1 else None
        type_ = date_type[0].strip()
        description = df_cleaned.iloc[index + 1][1] if index + 1 < len(df_cleaned) else None  # Next row's description
        paid_in = df_cleaned.iloc[index + 1][2] if index + 1 < len(df_cleaned) else None  # Next row's Paid In
        paid_out = df_cleaned.iloc[index + 1][3] if index + 1 < len(df_cleaned) else None  # Next row's Paid Out
        balance = df_cleaned.iloc[index + 1][4] if index + 1 < len(df_cleaned) else None  # Next row's Balance

        structured_data.append([date, type_, description, paid_in, paid_out, balance])

# Convert to DataFrame
final_df = pd.DataFrame(structured_data, columns=['Date', 'Type', 'Description', 'Paid in', 'Paid out', 'Balance'])

# Display the final structured DataFrame
print(final_df)
```
