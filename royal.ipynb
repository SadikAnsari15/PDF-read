```python
import camelot
import pandas as pd

# Extract tables from all pages using the 'stream' flavor
tables = camelot.read_pdf("royal-bank-statement-1.pdf", pages="all", flavor="stream")

# Combine all tables into a single DataFrame
dfs = [table.df for table in tables]
df = pd.concat(dfs, ignore_index=True)

# Drop any empty rows or columns if needed
df_cleaned = df.dropna(how='all', axis=0)
df_cleaned = df_cleaned.dropna(how='all', axis=1)  # Optionally drop empty columns

# Reset index and print the cleaned DataFrame
df_cleaned.reset_index(drop=True, inplace=True)
print(df_cleaned)
```

```python
df_cleaned = df_cleaned[df_cleaned[1].notna() & (df_cleaned[1] != "")]
# Extract data starting from the desired row (row 16 here)
#df_cleaned = df_cleaned.iloc[16:].reset_index(drop=True)
df_cleaned.head(35)
```

```python
df_cleaned.columns = df_cleaned.iloc[0]  # Set the first row as columns
df = df_cleaned[1:]  # Remove the first row after setting it as the header
```

```python
df_cleaned.columns = df_cleaned.iloc[0]  # Set the first row as columns
df = df_cleaned[1:]  # Remove the first row after setting it as the header
# Define new column names
new_column_names = ["Date", "Description","Type", "Credits", "Debits"]
 
# Assign new column names to the DataFrame
df.columns = new_column_names
```

```python
# Remove rows where 'Date' column value is "Date"
df = df[df['Date'] != "Date"]

# Reset index again after filtering
df.reset_index(drop=True, inplace=True)
```
